```python
import numpy as np
import matplotlib.pyplot as plt 

def observe():
    return np.mean(np.random.choice([0,1], 20, [0.9, 0.1]))

N = 1000000
alpha = np.arange(0.1,1.05,step=0.05)
fre = []
for a in alpha:
    fre.append(np.mean([observe()>a for i in range(N)]))
```

```python
EX = 0.5
DX = 0

plt.plot(alpha, fre, label='frequency')
plt.plot(alpha, [1/(10*a) for a in alpha], label="Markov's bound")
plt.plot(alpha, [9/(100*(a-0.1)**2) if a != 0.1 else 50 for a in alpha], label="Chebyshev's bound")
plt.plot(alpha, [np.exp(-40*(a-0.1)**2) if a != 0.1 else 50 for a in alpha], label="Hoeffding's bound")

plt.legend()
plt.xlabel('alpha')
plt.ylabel('frequency')
```
